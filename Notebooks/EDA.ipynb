```python
import pandas as pd
```

```python
import numpy as np
```

```python
df=pd.read_csv("clean_sp100_data.csv")
```

```python
df.head()
```

```python
aapl_vol = df[['Date', 'AAPL_20d Volatility']].dropna()
```

```python
aapl_vol.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))
plt.plot(aapl_vol['Date'], aapl_vol['AAPL_20d Volatility'])
plt.title('AAPL 20-Day Rolling Volatility')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.show()
```
