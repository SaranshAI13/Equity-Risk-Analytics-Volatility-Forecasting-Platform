```python
raw = pd.read_csv("sp100_stocks_data.csv", header=None)
raw.head(6)
```

```python
tickers = raw.iloc[0]
fields  = raw.iloc[1]
```

```python
final_cols = ['Date']   # Date forcefully first
last_field = None

for t, f in zip(tickers[1:], fields[1:]):  # column 1 se start
    if pd.notna(f):
        last_field = f

    if pd.isna(t) or pd.isna(last_field):
        final_cols.append(None)
    else:
        final_cols.append(f"{t}_{last_field}")
```

```python
data = raw.iloc[4:].copy()
data.columns = final_cols
```

```python
data = data.loc[:, data.columns.notna()]
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data = data.sort_values('Date').reset_index(drop=True)
```
